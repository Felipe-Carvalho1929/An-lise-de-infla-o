O principio deste trabalho visa análisar da inflação na região da Ásia
Para tanto será estudado a média, máxima, mínima, ano e região com maior inflação. Após isso será feita uma análise especifica para cada região 

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

Importar bibliotecas e criar as listas com as regiões: 
Caucaso e ásia central,
Leste asiático,
Sul da ásia,
Sudeste asiático.

```python
data_Base = pd.read_csv('/kaggle/input/inflation-rate-in-asia/ADO April 2023 - Inflation.csv', encoding='latin-1')

Região = ['Caucasus and Central Asia','East Asia','South Asia','Southeast Asia']
Cau = ['Kazakhstan','Pakistan', 'Turkmenistan', 'Uzbekistan', 'Tajikistan' 'Armenia','Azerbaijan','Georgia','Afghanistan']
leste = ['China','Republic of Korea','Hong Kong','Singapore','Indonesia', 'Malaysia', 'Philippines', 'Thailand']
sul = ['Bangladesh','India','Pakistan','Bhutan','Sri Lanka','Nepal' 'Maldives',]
sudeste = ['Brunei Darussalam', 'Cambodia','Indonesia', 'Malaysia', 'Philippines','Lao Peoples Dem. Rep.','Myanmar','Singapore' ,'Thailand', 'Timor-Leste','Viet Nam']

#data = data_Base[~data_Base['RegionalMember'].isin(outras)]
By_region = data_Base[data_Base['RegionalMember'].isin(Região)]
By_Caucaso_central = data_Base[data_Base['RegionalMember'].isin(Cau)]
By_lest = data_Base[data_Base['RegionalMember'].isin(leste)]
By_south = data_Base[data_Base['RegionalMember'].isin(sul)]
By_southeast = data_Base[data_Base['RegionalMember'].isin(sudeste)]
```

```python
regioes_unicas = By_region['RegionalMember'].unique()

# Imprimir as regiões
for regiao in regioes_unicas:
    print(regiao)
```

**Primeiramente será feito uma análise por região**

```python
    # Região com maior inflação no período
    regiao_maior_inflacao1 = By_region.groupby('RegionalMember')['Inflation'].sum().idxmax()
    print(f"Região com maior inflação no período: {regiao_maior_inflacao1}")
   # Região com menor inflação no período
    regiao_menor_inflacao1 = By_region.groupby('RegionalMember')['Inflation'].sum().idxmin()
    print(f"Região com menor inflação no período: {regiao_menor_inflacao1}")
    # 2. Identificar o ano de maior e menor inflação por região
    ano_maior_inflacao1 = By_region.groupby('Year')['Inflation'].sum().idxmax()
    ano_menor_inflacao1 = By_region.groupby('Year')['Inflation'].sum().idxmin()
    print(f"Ano de maior inflação: {ano_maior_inflacao1}")
    print(f"Ano de menor inflação: {ano_menor_inflacao1}")
```

Em uma primeira análise temos que a região de maior inflação no periodo é Caucasus e Asia Central. Por outro lado a região com menor inflação e o leste asiático.
Podemos observar também que em 2018 (inicio da análise) foi o periodo com a menor inflação na região. Em 2022, possivelmente decorrente da crise proporcionada pelo Covid 19, tem-se o periodo de maior inflação.

```python
 paises = By_region['RegionalMember'].unique()

for pais in paises:
    df_pais = By_region[By_region['RegionalMember'] == pais]
    plt.plot(df_pais['Year'], df_pais['Inflation'], label=pais)

    plt.xlabel('Ano')
    plt.ylabel('Inflação')
    plt.title(f'Inflação no país {pais}')
    plt.legend()
    plt.show()
```
